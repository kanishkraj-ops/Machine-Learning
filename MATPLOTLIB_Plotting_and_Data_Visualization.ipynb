# **INTRO TO MATPLOTLIB**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot();
#add a semicolon after this command so that you dont get the object
# reference in sqaure brackets , you can ignore it if you like the other way
```

```python
x = [1,2,3,4]
y = [4,5,6,7]
plt.plot(x,y);
```

```python
#1st method
fig = plt.figure() #creates a figure that is the a blank canvas
ax = fig.add_subplot() #creates a drawing area an axes to the canvas
ax.plot(x,y);
```

```python
#2nd method(recommended)
fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
#3rd method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y);
```

```python
import numpy as np
```

```python
#create some data
x = np.linspace(0,10,1000)
y = x**2
```

```python
#Make plot from dictionary
car_prices = {
    "BMW":10000,
    "AUDI" :35004,
    "MERCEDES":33440,
}
fig, ax = plt.subplots()
ax.bar(car_prices.keys(),car_prices.values());
ax.set(title='Car Dealership sales',
       xlabel = 'Car',
       ylabel='Price($)');
```

```python
fig,ax = plt.subplots()
ax.barh(list(car_prices.keys()),list(car_prices.values()));
```

```python
#subplots
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax1.plot(x,y)
ax2.scatter(x,y)
ax3.bar(car_prices.keys(),car_prices.values());
ax4.hist(np.random.randn(1000));
```

##**PLOTTING FROM PANDAS DATAFRAME**

```python
car_sales = pd.read_csv("sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[$\,\.]" , "" ,regex=True)
car_sales['Price'] = car_sales['Price'].astype(int)
car_sales
```

```python
car_sales['Price'] = car_sales['Price']/100
car_sales
```

```python
car_sales["sale_date"] = pd.date_range("2021-01-01",periods=len(car_sales))
car_sales
```

```python
car_sales["Total_sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(x="sale_date",y="Total_sales");
```

```python
car_sales.plot(x="Odometer (KM)",y="Price",kind="scatter");
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease.plot()
```

```python
heart_disease.plot.hist(subplots=True, figsize=(5,15));
```

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50
```

```python
over_50.plot(x='age',
             y='chol',
             c='target',
             kind='scatter');
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax = ax
            );
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],y=over_50['chol'],c=over_50['target'])
ax.set(title='Heart Disease and Cholestrol Levels',xlabel='Age',ylabel='Cholestrol');
#add a legend or you can say scale

ax.legend(*scatter.legend_elements(),title='Target');
#add a horizontal line
ax.axhline(over_50['chol'].mean(),color='red',linestyle='--');
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["chol"],y=over_50['thalach'],c=over_50['target']);
ax.set(title='Heart Disease and Cholestrol Levels',xlabel='Cholestrol',ylabel='Heart-Rate');
#add a legend or you can say scale

ax.legend(*scatter.legend_elements(),title='Target');
#add a horizontal line
ax.axhline(over_50['thalach'].mean(),color='red',linestyle='--');
```

```python
fig, (ax0,ax1) = plt.subplots(ncols=1,nrows=2,figsize=(10,10))

#add data to ax0
scatter = ax0.scatter(x=over_50["chol"],y=over_50['thalach'],c=over_50['target']);
ax0.set(title='Heart Disease and Cholestrol Levels',xlabel='Cholestrol',ylabel='Heart-Rate');
#add a legend or you can say scale

ax0.legend(*scatter.legend_elements(),title='Target');
ax0.axhline(y=over_50['thalach'].mean(),color='red',linestyle='--');



#add data to ax1

scatter = ax1.scatter(x=over_50["age"],y=over_50['thalach'],c=over_50['target']);
ax1.set(title='Heart Disease and Cholestrol Levels',xlabel='Age',ylabel='Heart-Rate');
#add a legend or you can say scale

ax1.legend(*scatter.legend_elements(),title='Target');
#add a horizontal line
ax1.axhline(over_50['thalach'].mean(),color='red',linestyle='--');
fig.suptitle('Heart Disease Analysis',fontweight='bold');
```
