# **INTRO TO MATPLOTLIB**

```python
%matplotlib inline #we can see the plots in our notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot();
#add a semicolon after this command so that you dont get the object
# reference in sqaure brackets , you can ignore it if you like the other way
```

```python
x = [1,2,3,4]
y = [4,5,6,7]
plt.plot(x,y);
```

```python
#1st method
fig = plt.figure() #creates a figure that is the a blank canvas
ax = fig.add_subplot() #creates a drawing area an axes to the canvas
ax.plot(x,y);
```

```python
#2nd method(recommended)
fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
#3rd method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y);
```

```python
#step-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#step-2
x=[1,2,3,4]
y=[10,35,33,67]

#step-3
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x,y)

#step-4
ax.set(title="Example plot",
       xlabel = 'x-axis',
       ylabel = 'y-axis')
#step-5
fig.savefig("plot.png")
```

# **DIFFERNT KINDS OF PLOTS**


*   HISTOGRAM
*   SCATTER PLOT
*   BAR PLOT
*   SUBPLOTS
*   LINE PLOTS






```python
import numpy as np
```

```python
#create some data
x = np.linspace(0,10,1000)
y = x**2
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
fig, ax = plt.subplots()
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots()
ax.scatter(x,np.log10(x));
```

```python
#Make plot from dictionary
car_prices = {
    "BMW":10000,
    "AUDI" :35004,
    "MERCEDES":33440,
}
fig, ax = plt.subplots()
ax.bar(car_prices.keys(),car_prices.values());
ax.set(title='Car Dealership sales',
       xlabel = 'Car',
       ylabel='Price($)');
```
