# **MANIPULATING DATA**

```python
#to lowercase or uppercase
car_sales['Make'] = car_sales["Make"].str.lower() # we also need to reassign
car_sales
```

 we can't use inplace=True with string accessor methods like .str.upper() in pandas.

Why?
Because methods under .str (like .str.upper(), .str.lower(), .str.replace(), etc.) return a new Series and do not support the inplace parameter.

```python
#or we can change inplace = True it will do the change inplace
car_sales["Make"] = car_sales["Make"].str.upper()
car_sales
```

# Handling Missing Data

```python
missing_sales = pd.read_csv("car-sales-missing-data.csv")
missing_sales
```

NaN means a non existent values and its pretty common to encounter NaN when dealing with real datasets.

```python
#But we dont want NaN in our data as it will interfere so we will handle it
missing_sales.isna()  #for checking if missing data is there, returns true for that
```

```python
missing_sales['Odometer'] = missing_sales['Odometer'].fillna(missing_sales['Odometer'].mean())
```

```python
missing_sales #fillna :- fills the NaN value with given value
```

```python
missing_sales.dropna(inplace = True) #drops all the rows with NaN
missing_sales
```

# **CREATING A NEW COLUMN**

```python
#from pandas series
seats = pd.Series([2,3,4,4,5,5,6,5])
car_sales['seats'] = seats
car_sales
```

```python
car_sales['seats'].fillna(5, inplace=True)
car_sales
```

```python
#from python list
#but remember creating a column from list isn't same as series as all 10 values needs to be present i.e dor all indices
fuel_economy = [7.5,9.2,5.0,9.6,8.7,4.7,7.6,8.7,3.0,4.5]
car_sales['fuel per 100KM'] = fuel_economy
car_sales
```

```python
#generating a colummn from another column
car_sales['total fuel used(L)'] = car_sales['Odometer (KM)']/100 * car_sales['fuel per 100KM']
car_sales
```

```python
#from a single number
car_sales['Number of wheels'] = 4
car_sales
```

```python
#Deleting/Dropping a column
car_sales = car_sales.drop("total fuel used" , axis=1)
car_sales
```

```python
#Shuffling the data
shuffeld_data = car_sales.sample(frac=1)
shuffeld_data
```

```python
#To reset the dataset we can use reset_index
shuffeld_data.reset_index(inplace=True)
shuffeld_data #we need to have drop=true to remove an extra index column that would appear
```

```python
shuffeld_data.reset_index(drop=True , inplace=True)
shuffeld_data
```

```python
#we can use .apply() to apply a function to a column
car_sales['Odometer (KM)'] = car_sales['Odometer (KM)'].apply(lambda x: x / 1.6093)
car_sales
```
